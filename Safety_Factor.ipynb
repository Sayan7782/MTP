```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[0.87,0.91,0.98,0.92,1.43,0.87,1.13,1.19]
y=[5.52,6.05,4.79,4.62,5.43,4.86,5.81,5.12]
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 5))

myline = np.linspace(0.7,1.5,100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.xlabel("Safety Factor ")
plt.ylabel("Face Advance Rate Actual(m/d)")
plt.show() 
```

```python
from sklearn.metrics import r2_score
print(r2_score(y, mymodel(x)))
```

```python
print(mymodel.coeffs)
```
