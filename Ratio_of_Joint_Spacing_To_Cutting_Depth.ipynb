```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1.32,1.19,1.36,0.73,1.41,1.20,0.98,1.14]
y=[5.52,6.05,4.79,4.62,5.43,4.86,5.81,5.12]
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 2))

myline = np.linspace(0,1.5,100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.xlabel("Ratio of Joint Spacing to Cutting Depth ")
plt.ylabel("Face Advance Rate Actual(m/d)")
plt.show() 
```

```python
from sklearn.metrics import r2_score
print(r2_score(y, mymodel(x)))
```

```python
print(mymodel.coeffs)
```
