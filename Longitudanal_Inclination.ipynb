```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[14.94,13.62,16.13,18.97,12.13,25.36,16.78,17.06]
y=[5.52,6.05,4.79,4.62,5.43,4.86,5.81,5.12]
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 2))

myline = np.linspace(5,30,100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.xlabel("Longitudanal Inclination ")
plt.ylabel("Face Advance Rate Actual(m/d)")
plt.show() 
```

```python
from sklearn.metrics import r2_score
print(r2_score(y, mymodel(x)))
```

```python
print(mymodel.coeffs)
```
