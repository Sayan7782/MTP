```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[154,148,150,126,123]
y=[40,38,41,37,39]
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 2))

myline = np.linspace(120, 160, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.xlabel("Face length(in m)")
plt.ylabel("Equipment Salvage(in m)")
plt.show() 
```

```python
from sklearn.metrics import r2_score
print(r2_score(y, mymodel(x)))
```

```python
print(mymodel.coeffs)
```
